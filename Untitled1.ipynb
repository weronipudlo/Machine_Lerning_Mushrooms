```python
#usuwanie zmiennych kategorycznych
le = LabelEncoder() #instancja klasy, która służy do kodowania etykiet klas
data=data.apply(LabelEncoder().fit_transform) # metoda fit_transform koduje etykiety klas i przekształca wartości w kolumnach na liczby całkowite
data['class'] = le.fit_transform(data['class']) #przypisanie etykiet klas do kolumny 'class' w postaci liczb całkowitych
data.head() #początkowe 5 wierszy
```

```python
#histogramy wszystkich zmiennych
data.hist(figsize=(15,15))
```

```python
plt.figure(figsize = (6,6))
plt.pie(data['class'].value_counts(), startangle = 90, autopct = '%.1f', labels = ['EDIBLE', 'POISONOUS'], shadow = True)
plt.show()
```

```python
X = data.drop(['class'], axis = 1) #usuwamy kolumne class i zwracamy nowy DataFrame z danymi objaśniającymi
Y = data['class'] #tworzymy serię zawierającą tylko klase grzyba - klasę, którą próbujemy przewidzieć
#dzielimy zbiór na zbiór treningowy - 75% i testowy - 25% | random_state - ustawienie wartości ziarna generatora liczb losowych, gwarantuje powtarzalność wyników
#X_train i X_test zawierają zmienne objaśniające, a Y_train i Y_test zawierają etykiety klas dla odpowiednio zbiorów treningowego i testowego
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25, random_state=0)
```

```python
X.head() #wgląd w zbiór danych (5 pierwszych danych)
```

```python
braki = data.isnull().sum()
print(braki)
```

```python
# wyświetlenie liczby wierszy i kolumn w zbiorach danych treningowych i testowych
print("X_test shape:",X_test.shape)
print("X_train shape:",X_train.shape)
```

```python
print("Y_test shape:",Y_test.shape)
print("Y_train shape:",Y_train.shape)
```

```python
model = tf.keras.Sequential() #tworzy instancję sekwencyjnego modelu w TensorFlow
model.add(tf.keras.layers.Input(shape=X_train.shape[1])) #warstwa wejściowa, warstwa wejściowa zostanie dostosowana do liczby cech w danych treningowych, które będą przekazywane do modelu
model.add(tf.keras.layers.Dense(600, activation="relu"))#funckje aktywacji ReLU
model.add(tf.keras.layers.Dense(300, activation="relu"))
model.add(tf.keras.layers.Dense(100, activation="relu"))
model.add(tf.keras.layers.Dense(2, activation="softmax"))
```

```python
model.compile(
    loss=keras.losses.sparse_categorical_crossentropy, #funkcja straty - Odpowiada za mierzenie błędu lub odległości pomiędzy predykcją modelu a rzeczywistymi danymi
    optimizer=keras.optimizers.Adam(), #optymalizator - minimalizuje f. straty - dostosowuje parametry do optymalnego zestawu wartości
    metrics=['sparse_categorical_accuracy'] #metryka
)
#model gotowy do proocesu uczenia
```

```python
model.summary()
```

```python
hist=model.fit(x=X_train, y=Y_train, epochs=40, validation_data=(X_test, Y_test))
```

```python
test_loss, test_acc = model.evaluate(X_test, Y_test)
print('Test accuracy', test_acc)
print('Test loss', test_loss)
```

```python
# Treningowe metryki
#train_acc = hist.history['accuracy']
#train_prec = hist.history['precision']
train_cat_acc = hist.history['sparse_categorical_accuracy']
train_loss = hist.history['loss']

# Metryki testowe
#test_acc= hist.history['val_accuracy']
test_cat_acc= hist.history['val_sparse_categorical_accuracy']
test_loss = hist.history['val_loss']

# Numer epoki
#epochs = range(1, len(train_cat_acc) + 1)
epochs = range(1, len(train_cat_acc) + 1)

# Wykresy
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(epochs, train_cat_acc, 'b', label='Dokładność k. treningowa')
plt.plot(epochs, test_cat_acc, 'r', label='Dokładność k. testowa')
plt.title('Dokładność k. treningu i testu')
plt.xlabel('Epoka')
plt.ylabel('Dokładność k.')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs, train_loss, 'b', label='Strata treningowa')
plt.plot(epochs, test_loss, 'r', label='Strata testowa')
plt.title('Strata treningu i testu')
plt.xlabel('Epoka')
plt.ylabel('Strata')
plt.legend()

plt.tight_layout()
plt.show()
```
